```python
from sklearn.model_selection import train_test_split


X_train_df, X_test_df , Y_train_df, Y_test_df = train_test_split(df, Y, test_size=0.25, random_state=66)
```

```python
X_train_df.shape, Y_train_df.shape
```

```python
from sklearn import preprocessing

# standard_scaler_num = preprocessing.StandardScaler()
# x_train_sc = standard_scaler_num.fit_transform(X_train_df)
# x_test_sc = standard_scaler_num.transform(X_test_df)
```

```python
data_norm = X_train_df.copy() #Has training + test data frames combined to form single data frame
normalizer = preprocessing.StandardScaler()
data_array = normalizer.fit_transform(data_norm)
X_train_sc_df = pd.DataFrame(data_array,columns = data_norm.columns).set_index(data_norm.index)


data_norm_test = X_test_df.copy()
data_array_test = normalizer.transform(data_norm_test)
X_test_sc_df = pd.DataFrame(data_array_test,columns = data_norm_test.columns).set_index(data_norm_test.index)
```

```python
pickle.dump(normalizer, open('standard_scalar.pkl', 'wb'))
```

### Checking for the constant variance columns

```python
from sklearn.feature_selection import VarianceThreshold
var_thresh = VarianceThreshold(threshold=0)

var_thresh.fit(X_train_sc_df)
```

```python
var_thresh.get_support()
```

```python
constant_cols = [col for col in X_train_sc_df.columns 
                if col not in X_train_sc_df.columns[var_thresh.get_support()]]
constant_cols
```

```python
# print(X.corr()) 
corr = pd.DataFrame(data_norm).corr(method='pearson')
plt.figure(figsize=(20,15))
# plotting correlation heatmap 
plt.title("Pearson correlation between the normalised features")
matrix = np.triu(corr)

# using the upper triangle matrix as mask 
sns.heatmap(corr, annot=True, mask=matrix, cmap="YlGnBu")
# dataplot = sns.heatmap(, cmap="YlGnBu", annot=True) 
  
# displaying heatmap 
plt.show() 
```

```python
X_train_df.columns
```

```python
corr_matrix = corr.abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(abs(upper[column]) > 0.80)]
print(to_drop)
# Drop features 
X_train_sc_df = X_train_sc_df.drop(to_drop, axis=1)
X_test_sc_df = X_test_sc_df.drop(to_drop, axis=1)
```

```python
X_test_df.iloc[0]
```

## Data modelling 

### Feature importances 

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=5)
x = X_train_sc_df
y = Y_train_df

clf.fit(x,y)
feature_imp = pd.DataFrame(clf.feature_importances_,index=x.columns)
feature_imp.sort_values(by = 0 , ascending = False)
```

```python
from sklearn.feature_selection import RFE
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
y_pred = clf.predict(X_test_sc_df)

print(classification_report(y_pred, Y_test_df))
```

```python
pickle.dump(clf, open("RC_clf.pkl", 'wb'))
```
